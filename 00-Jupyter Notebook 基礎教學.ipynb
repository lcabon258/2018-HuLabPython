# 使用 Jupyter Notebook 優勢：
+ 互動開發，寫一段執行一段，同時做筆記。
+ 多種語言支援，除了Python以外，也有[多種核心（語言）](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels)，如 R 語言等。

# 區塊種類：
+ Code：程式碼，可以互動式執行（RShift + Enter/Return）
+ Markdown：使用 Markdown 語言編寫說明，簡單編寫語言架構，也可以加入html語法做到更高程度的客製化。

# Markdown 入門：
Markdown 是一個讓使用者可以快速編寫的語言，使用標點符號（如：\[ \] \* \# \+ ）簡化複雜的格式設定，實際呈現時，再轉換為 HTML 語法，核心價值為「易讀易寫」。

更詳細的說明可以參考[這邊](https://markdown.tw/)或是[這邊](https://guides.github.com/features/mastering-markdown/)，下面將介紹比較常用的一些語法。

歐對了，你可以在使用 Notebook 時**點兩下 Markdown 的區塊**，就可以開始編輯原始文字囉！

# 清單：使用「\*」、「\+」、「-」或數字
**無序清單**
使用時把```*或+或-```後面空一個，即可變成清單，每空四格，即可增加一階。
比如：  

---
```markdown
+ Japan
+ Taiwan
    + Taipei
        + NTU
        + NTNU
```  

---

顯示起來會向這樣：  

---  
+ Japan
+ Taiwan
    + Taipei
        + NTU
        + NTNU

---
要注意的是：清單也可以有多個段落，**只要前面的空格數目相同，就會被歸類在同一項裡面**，比如：

---
```markdown
+ 顯示器  
    + Liquid Retina HD 顯示器  
    6.1 吋 (對角線) 全螢幕 LCD 多點觸控顯示器，採用  
    IPS 技術
    *1792 x 828 像素，326 ppi 解析度*  
```

---
顯示：

---
+ 顯示器  
    + Liquid Retina HD 顯示器  
    6.1 吋 (對角線) 全螢幕 LCD 多點觸控顯示器，採用  
    IPS 技術
    *1792 x 828 像素，326 ppi 解析度*  
---
小問題：為什麼「*1792 x 828 像素，326 ppi 解析度*」跑到上面了呢？

**有序清單**
有序清單非常簡單，就是把「數字」加上「\.」即可！層級設定方式也是依照前方空格數量決定。
比如：

---
```markdown
1. 象山
2. 台北101
3. 信義安和
```

---

顯示為：  

---
1. 象山  
2. 台北101  
3. 信義安和  
---  
有個神奇的小玩意：就是前面數字的順序是不重要的！  

---
```markdown
3. 象山
49. 台北101
9. 信義安和
```  

---  

顯示為：  

---
3. 象山
49. 台北101
9. 信義安和

# 強調語法：

比如說：  

---

---  
我是*比薩斜塔*  
_好斜好斜_

```python
def bubble_sorted(iterable):
    new_list = list(iterable)
    list_len = len(new_list)
    for i in range(list_len - 1):
        for j in range(list_len - 1, i, -1):
            if new_list[j] < new_list[j - 1]:
                new_list[j], new_list[j - 1] = new_list[j - 1], new_list[j]
    return new_list
```

````python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
````

````python
def plus(x,y):
    return x+y
x_w = widgets.IntSlider(value=5,min=0,max=10,step=1,description="X=")
y_w = widgets.IntSlider(value=5,min=0,max=10,step=1,description="Y=")
interact(plus,x=x_w,y=y_w)
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

````python
def f(m, b):
    """示範如何顯示 y = mx + b"""
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
````
